```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('Train.csv').values
```

```python
X=data[:,:-1]
Y=data[:,-1]
```

```python
print(X[1][0])
```

# cleaning

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer=RegexpTokenizer('[a-zA-Z]+')
ps=PorterStemmer()
sw=set(stopwords.words('english'))
```

```python
def myTokenizer(sentence):
    sentence=sentence.replace("<br /><br />"," ")
    sentence = tokenizer.tokenize(sentence.lower())
    new_text= [ps.stem(w) for w in sentence if w not in sw]
    final=' '.join(new_text)
    return final
```

```python
for review in range(X.shape[0]):
    
    clean_review=myTokenizer(X[review][0])
    X[review]=clean_review
```

```python
X_clean = []
for z in X:
    zl=z.tolist()
    X_clean.append(zl[0])
X_clean
```

```python
X_clean[:2]
```

# vertorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range=(1,2))

x_vec = cv.fit_transform(X_clean)
```

```python
print(x_vec)
```

```python
a=cv.vocabulary_.keys()
print(type(a))
```

# test input

```python
XT=pd.read_csv('Test.csv').values
```

```python
for review in range(XT.shape[0]):
    
    clean_review=myTokenizer(XT[review][0])
    XT[review]=clean_review
```
